```python
# Exercise 22: Remove common items from array
a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 6, 7, 8, 9])

# c = np.in1d(a,b) New concept
# mask = np.in1d(a,b) in1d is deprecated.
# using np.isin()
mask = np.isin(a,b)
result = a[~mask] #`~` inverts the mask index :bitwise NOT
print(result)
```

```python
# Exercise 27: Sorting a NumPy array based on a specific column
sampleArray = np.array([[34, 43, 73], [82, 22, 12], [53, 94, 66]])

sorted_indices = sampleArray[:, 1].argsort()

sorted_array = sampleArray[sorted_indices]

print("OG Array: ", sampleArray)
print("Sorted Array: ", sorted_array)
```

```python
# Exercise 47: Create an 8×8 checkerboard pattern using 0s and 1s


checkerboard = np.zeros((8,8), dtype= int)

checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print(checkerboard)
```

```python
# Exercise 48: Find nearest value

arr = np.array([1.5, 2.8, 3.2, 4.1])
target_value = 7

index = np.abs(arr - target_value).argmin()
nearest_value = arr[index]

print(f"The value closest to {target_value} is: {nearest_value}")
```
