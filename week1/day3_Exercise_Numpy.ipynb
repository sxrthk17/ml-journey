```python
# Exercise 18: Find the common items between two arrays
a = np.array([1, 2, 3, 2, 8, 4, 2, 4])
b = np.array([2, 4, 5, 6, 8])

c = np.intersect1d(a,b)
print(c)
```

```python
# Exercise 21: Compute the mean, median, and standard deviation of a NumPy array
import numpy as np
arr = np.array([10, 20, 30, 100, 200, 300])

print("Mean: ", arr.mean())
print("Median: ",np.median(arr))
print("Standard deviation: ",arr.std())
```

```python
# Exercise 22: Remove common items from array
a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 6, 7, 8, 9])

# c = np.in1d(a,b) New concept
# mask = np.in1d(a,b) in1d is deprecated.
# using np.isin()
mask = np.isin(a,b)
result = a[~mask] #`~` inverts the mask index :bitwise NOT
print(result)
```

```python
# Exercise 24: Get the positions where elements of array a and b match
a = np.array([1, 2, 3, 4, 5])
b = np.array([1, 4, 3, 7, 8])

c = np.where(a == b)
print(c)
```
